```python
nltk.download(['brown','gutenberg','inaugural'])
```

```python
brown = nltk.corpus.brown
```

```python
#cfd
cfd = nltk.ConditionalFreqDist(
    (genre,word)
    for genre in brown.categories()
    for word in brown.words(categories = genre)
)

modals = ['can','could','would','will','might','may']
cfd.tabulate(conditions = brown.categories(), samples = modals)
```

```python
# modals의 빈도수가 포함 된 cfd (word -> modal_verb)
cfd1 = nltk.ConditionalFreqDist(
    (genre, modal)
    for genre in brown.categories()
    for modal in ['can','could','would','will','might','may']
    
    for word in brown.words(categories = genre)
        if word.lower() == modal
)

cfd1.tabulate()
```

- tabulate ()

- plot ()

```python
#순회 zip, enumerate
name = ['kate', 'royce','phill']
age = [77, 44, 54]

#(name, age)

for item1, item2 in zip(name, age): #여러 리스트를 동시에 순회
    print((item1, item2))
```

```python
for item1 in name:
    for item2 in age:
        print((item1, item2)) #기존 중첩for문
```

```python
list(zip(name, age))
```

```python
# list + index가 필요할 때: enumerate
for idx, item in enumerate(name): #이 줄은 순서 변경 X, 인덱스 먼저, 아이템
    print(idx,item)
```

```python
#리스트컴프리헨션으로 바꾸면

[word for word in zip(name, age)]
```

```python
#리스트컴프리헨션으로 바꾸면2 / 각각 변화 주고 싶으면 두개로 받으라고
[(i, j+33) for i, j in zip(name,age)]
```

```python
sentence = 'do not judge a book by its cover'

#[(대문자, 첫글자대문자, 글자수)] 리스트 만들기

sentence.split()
[(word.upper(),word.title(), len(word)) for word in sentence.split()]
```

```python
text = [
    ['Sunja', 'said', 'goodnight', 'to', 'her', 'mother'],
    ['Hansu', 'was', 'no', 'longer', 'in', 'Busan']
]

#1. 문장마다 str변환해서 리스트로 반환
# ['문장1', '문장2']

[' '.join(sent) for sent in text]
```

```python
#2. 전체를 스트링으로 변환
# '문장 1 문장 2'
[' '.join(' '.join(sent)) for sent in text]
```

```python
#3. 전체 단어 리스트로 가져오기
# [sunja, ssaid, goodnight, ......]
[word for sent in text for word in sent]
```
